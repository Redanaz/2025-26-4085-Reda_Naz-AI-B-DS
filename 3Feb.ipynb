```python
#List
superstore=["metro",'tesco','walmart','kmart','carrefour']
print(superstore)
number_list=[4,4,6,7,2,9,10,15]
mixed_list=['cat',87.23,65,[9,7,3,3]]
print(type(mixed_list))
#list operations
list1=[3,5,7,1]
list2=[2,4,6,8]
print(list1+list2) #list concartination
print(list1*3)
print(list1==list2)
#string to list
quote='Hello World'
string_to_list=list(quote)
print(string_to_list)
# list indexing, slicing
print(list1[:2])
print(list1[-2:])
print(5 in list1)
print(11 in list1)
#list built-in functions
cars=['BMW','Audi','Mercedes','Toyota']
print(len(cars))
print(max(number_list))
print(min(number_list))
print(sum(number_list))
print(sorted(list1))
print(sorted(list1,reverse=True))
#list methods
cars.append('Tesla')
print(cars)
cars.insert(2,'Ford')
print(cars)
cars.remove('Tesla')
print(cars)
cars.pop()
print(cars)
#Nested list
nested_list=[[1,2,3],[4,5,6],[7,8,9]]
print(nested_list)
print(nested_list[1])
print(nested_list[1][2])
```

```python
# Creating a set
fruits = {"apple", "banana", "cherry"}
print("Original Set:", fruits)
fruits.add("orange")
print("\nAfter add():", fruits)
fruits.update(["mango", "grapes"])
print("\nAfter update():", fruits)
fruits.remove("banana")
print("\nAfter remove('banana'):", fruits)
fruits.discard("pineapple")
print("\nAfter discard('pineapple'):", fruits)
removed_item = fruits.pop()
print("\nAfter pop():", fruits)
fruits_copy = fruits.copy()
print("\nCopied Set:", fruits_copy)
print("\nLength of set:", len(fruits))
# Checking membership
if "apple" in fruits:
    print("\n'apple' is present in the set")
A={1,2,3,4,5}
B={4,5,6,7,8}
print("\nUnion:", A.union(B))
print("Intersection:", A.intersection(B))
print("Difference (A - B):", A.difference(B))
print("Symmetric Difference:", A.symmetric_difference(B))
print("\nIs A subset of B?", A.issubset(B))
print("Is A superset of B?", A.issuperset(B))
fruits.clear()
print("\nAfter clear():", fruits)
```

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 22,
    "course": "Computer Science"
}

print("Original Dictionary:", student)
#Accessing values
print("\nAccessing Values:")
print("Name:", student["name"])
print("Age:", student.get("age"))
student["grade"] = "A"
print("\nAfter Adding grade:", student)
student["age"] = 23
print("\nAfter Updating age:", student)
student.pop("course")
print("\nAfter pop('course'):", student)
student.popitem()
print("\nAfter popitem():", student)
del student["age"]
print("\nAfter deleting age:", student)
print("\nLength of dictionary:", len(student))
print("\nKeys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())

#Looping through dictionary
print("\nLooping through dictionary:")
for key, value in student.items():
    print(key, ":", value)

student_copy = student.copy()
print("\nCopied Dictionary:", student_copy)
student.clear()
print("\nAfter clear():", student)
#Creating dictionary using fromkeys()
keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)
print("\nDictionary using fromkeys():", new_dict)
```
